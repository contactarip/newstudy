#### Cleaning Data
there are some wrong type on feature that values are typo and null values. So we will clear the data with thecnique replace the incorrect value, and than fillna to that null values.

Null Values : on feature "Amount" and "Individual_Price_US$".

Typo Values : on feature  "Product","Category" and "Delivery_Type".

```python
df["Amount US$"] = df["Amount US$"].fillna(0)
df["Individual_Price_US$"]= df["Individual_Price_US$"].fillna(0)
```

```python
df.Product.replace({"Jeans" : "Jean",
                     "Shirts" : "Shirt"}, inplace = True)
df.Category.replace({"clothing" : "Clothing"}, inplace = True)
df.Delivery_Type.replace({"one-day deliver" : "one-day delivery"}, inplace = True)
```
